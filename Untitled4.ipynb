## Check for anomalies or wrong entries

```python
numerical_columns = df.select_dtypes(include=[np.number]).columns
anomalies = {column: (df[column] < 0).sum() for column in numerical_columns if (df[column] < 0).any()}
anomalies
```

###Check for outliers (simple check using IQR method)

```python
outliers = {}
for column in numerical_columns:
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    outliers[column] = ((df[column] < (Q1 - 1.5 * IQR)) | (df[column] > (Q3 + 1.5 * IQR))).sum()
outliers
```

#Application and Enrollment Analysis

#####Average number of applications recieved by colleges

```python
average_applications = df['Apps'].mean()
average_applications
```

##### Percentage of applications accepted on average across all colleges

```python
acceptance_rate = (df['Accept'].sum() / df['Apps'].sum()) * 100
acceptance_rate
```

###### Average enrollment rate (number of students enrolled divided by number of applications accepted)

```python
enrollment_rate = (df['Enroll'].sum() / df['Accept'].sum()) * 100
enrollment_rate
```

##### College with the highest number of applications received

```python
highest_apps_college = df.loc[df['Apps'].idxmax(), 'Names']
highest_apps_college
```
